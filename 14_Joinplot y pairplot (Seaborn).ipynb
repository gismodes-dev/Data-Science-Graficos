```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

#Jointplot
muestra 2 graficas en 1 sola

```python
sns.jointplot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,  #marginal trcicks sirve para agregar una tabla extra al histograma
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

#pairplot
sirve para la relacion entre variables numericas

```python
sns.pairplot(data=tips,hue='sex')  #diagrama de kde (densidades)
plt.show
```

```python
sns.pairplot(data=tips,hue='sex',corner=True)  #diagrama de kde (densidades)
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
